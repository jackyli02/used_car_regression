## Exploratory Analysis

```python
# we only want to look at cars from 1980 to 2020
# these before 1980 are too old and thus won't make meaningful predictions
vehicles_yr_ftd = vehicles[vehicles['year'].between(1950.0, 2020.0)]
```

```python
# convert the year to type int
vehicles_yr_ftd['year'] = vehicles_yr_ftd['year'].apply(lambda x: int(x))
```

```python
# remove columns that has nothing to do with the car itself
vehicles_3 = vehicles_yr_ftd[['region', 'price', 'year', 'manufacturer', 'model',
                              'condition', 'cylinders', 'fuel', 'odometer',
                              'title_status', 'transmission', 'drive', 'size',
                              'type', 'paint_color', 'county','state', 'lat', 'long']]
```

```python
# price_manu = pd.concat([vehicles_6['price'], vehicles_6['manufacturer']], axis=1)
# f, ax = plt.subplots(figsize=(15, 6))

fig = sns.boxplot(vehicles_3['price'][vehicles_3['price'] < 1e5])
```

```python
# first we look at the dependent variable that we're trying to predict
# eliminating the outliers in price (we use the iqr here since the z score is too large)
q1 = vehicles_3['price'].quantile(0.25)
q3 = vehicles_3['price'].quantile(0.75)
iqr = q3 - q1
in_range = (vehicles_3['price'] > q1 - 1.5 * iqr) & (vehicles_3['price'] < q3 + 1.5 * iqr)
vehicles_4 = vehicles_3[in_range]
```

```python
# drop vehicles with price less or equal to 200 (based on research, cars at these prices aren't meaningful listings)
vehicles_5 = vehicles_4[vehicles_4['price'] > 200]
```

```python
sns.set_style("white")
f, ax = plt.subplots()
sns.distplot(vehicles_5['price'])
ax.set(ylabel="Frequency")
ax.set(xlabel="Price")
ax.set(title="Price distribution")
plt.show()
vehicles_5['price'].describe()
```

```python
# Fill missing values
def percent_missing(df):
    data = pd.DataFrame(df)
    df_cols = list(pd.DataFrame(data))
    dict_x = {}
    for i in range(0, len(df_cols)):
        dict_x.update({df_cols[i]: round(data[df_cols[i]].isnull().mean()*100,2)})
    
    return dict_x
df_miss = sorted(percent_missing(vehicles_5).items(), key=lambda x: x[1], reverse=True)
df_miss
```

```python
vehicles_6 = vehicles_5.drop('county', axis = 1)
```

```python
def handle_missing_values(features):
    features['size'] = features['size'].fillna(features['size'].mode()[0])
    features['condition'] = features['condition'].fillna(features['condition'].mode()[0])
    features['cylinders'] = features['cylinders'].fillna(features['cylinders'].mode()[0])
    features['paint_color'] = features['paint_color'].fillna(features['paint_color'].mode()[0])
    features['drive'] = features['drive'].fillna(features['drive'].mode()[0])
    features['type'] = features['type'].fillna(features['type'].mode()[0])
    features['odometer'] = features['odometer'].fillna(round(vehicles_6['odometer'].mean(), 0))
    features['manufacturer'] = features['manufacturer'].fillna(features['manufacturer'].mode()[0])
    features['model'] = features['model'].fillna(features['model'].mode()[0])
    features['lat'] = features['lat'].fillna(round(vehicles_6['lat'].mean(), 4))
    features['long'] = features['long'].fillna(round(vehicles_6['long'].mean(), 4))
    features['fuel'] = features['fuel'].fillna(features['fuel'].mode()[0])
    features['title_status'] = features['title_status'].fillna(features['title_status'].mode()[0])
    features['transmission'] = features['transmission'].fillna(features['transmission'].mode()[0])
```

## Model Building

```python
vehicles_6.dtypes
```

```python
len(vehicles_6['region'].unique())
```

let's now consider the states because there would be too many '0' s in certain columns

```python
# get dummy variables
vehicles_7 = vehicles_6[['price', 'year', 'cylinders', 'odometer', 'manufacturer', 'condition',
                         'fuel', 'title_status', 'transmission','drive', 'size', 'type', 'paint_color']]
vehicles_7_dum = pd.get_dummies(vehicles_7)
```

```python
vehicles_7_dum.shape
```

```python
# split
x = vehicles_7_dum.drop('price', axis = 1)
y = vehicles_7_dum['price'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Standardize the data
scaling = StandardScaler()
x_train[['year', 'cylinders', 'odometer']] = scaling.fit_transform(x_train[['year', 'cylinders', 'odometer']])
x_test[['year', 'cylinders', 'odometer']] = scaling.fit_transform(x_test[['year', 'cylinders', 'odometer']])
```

```python
lm = LinearRegression()
lm.fit(X = x_train, y = y_train)
```

```python
np.mean(cross_val_score(lm, x_train, y_train, scoring = 'neg_mean_absolute_error', cv=3))
```

```python
lm.score(x_train, y_train)
```

```python
lm.coef_
```

```python
ridge = Ridge(alpha = 1)
ridge.fit(x_train, y_train)
ridge.coef_
ridge.score(x_train, y_train)
```

```python
cvs_r_stand = []
alp = np.array([1e-7, 0.001, 0.01, 0.05, 0.1, 1, 5, 10, 100])
for i in alp:
    ridge = Ridge(alpha = i)
    ridge.fit(x_train, y_train)
    cvs_r_stand.append(np.mean(cross_val_score(ridge, x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3)))
cvs_r_stand
```

```python
lasso = Lasso(alpha = 0.001)
lasso.fit(x_train, y_train)
lasso.coef_
lasso.score(x_train, y_train)
```

```python
cvs_l_stand = []
alp = np.array([1e-7, 0.001, 0.01, 0.05, 0.1, 1, 5, 10])
for i in alp:
    lasso = Lasso(alpha = i)
    lasso.fit(x_train, y_train)
    cvs_l_stand.append(np.mean(cross_val_score(ridge, x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3)))
cvs_l_stand
```

Neither of the above models worked very well (low R values)

```python
model = RandomForestRegressor(random_state=1)
model.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

Here, by using a random forest regressor, we achieve an accuracy of 75.98%.
